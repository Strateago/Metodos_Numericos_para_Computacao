**Questão 10 (letra b):  
y’’ + y = cos(t) com y(0) = y'(0) = 0**



Conforme foi resolvido na lista, obtemos como solução:
\begin{equation}
yp = \frac{tsen(t)}{2}
\end{equation}
E para essa equação temos que:
\begin{equation}
yp(0) = 0
\end{equation}
E:
\begin{equation}
yp' = \frac{tcos(t)+sen(t)}{2} => yp'(0) = 0
\end{equation}
O que atende as condições iniciais.

Vamos verificar a solução obtida numericamente (Pelo python):
\begin{equation}
y'' + y = cos(t) => y'' = cos(t) - y
\end{equation}

```python
#Mostrar o gráfico de y'' = cos(t) - y
import numpy as np;
from scipy.integrate import odeint
import matplotlib.pyplot as plt;

def function(y, t):
  d2ydt2 = np.cos(t) - y[0]
  return y[1], d2ydt2

y0 = [0, 0]
t = np.linspace(0, 10)
X = odeint(function, y0, t)
Y = X[:,0]
plt.plot(t, Y, 'r')
plt.xlabel('t');
plt.ylabel('y(t)');
```
