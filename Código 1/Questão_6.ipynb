**Questão 6:  
y(0) = 1**

Parte 1: *dy/dt = (y+4) cos(t)*:

\begin{equation}
y'= (y+4) cos(t)
\end{equation}

Utilizando o formato de equações separáveis:

\begin{equation}
y' = \frac{g(t)}{f(y)}
\end{equation}

Temos:

\begin{equation}
g(t) = cos(t)
\end{equation}

\begin{equation}
f(y) = \frac{1}{y+4}
\end{equation}

Podemos encontrar a solução *y* fazendo:

\begin{equation}
\int_{t0}^{t} g(t) dt = \int_{y0}^{y} f(y) dy
\end{equation}

Como queremos para y(0) = 1, temos que t0 = 0 e y(0) = 1, e substituindo as funções:

\begin{equation}
\int_{0}^{t} cos(t)dt = \int_{1}^{y} \frac{1}{y+4}dy
\end{equation}

Resolvendo as integrais temos:

\begin{equation}
sen(t) = ln(\frac{y+4}{5})
\end{equation}

Exponenciando de ambos os lados:

\begin{equation}
e^{sen(t)} = \frac{y+4}{5}
\end{equation}

\begin{equation}
5 e^{sen(t)} = y + 4
\end{equation}

E finalmente:

\begin{equation}
y = 5 e^{sen(t)} - 4
\end{equation}

Vamos verificar o gráfico para a solução calculada numericamente em python da equação diferencial:

\begin{equation}
y' = (y+4)cos(t)
\end{equation}

```python
#Mostrar o gráfico de y' = (y+4) cos(t)
import numpy as np;
from scipy.integrate import odeint
import matplotlib.pyplot as plt;

def function(y, t):
  dydt = (y+4)*np.cos(t)
  return dydt

y0 = 1
t = np.linspace(0, 5)
Y = odeint(function, y0, t)
plt.plot(t, Y)
plt.xlabel('t');
plt.ylabel('y(t)');
```

Vamos verificar agora o grafico da solução obtida analiticamente:

\begin{equation}
y = 5e^{sen(t)}-4
\end{equation}

```python
#Mostrar o gráfico da solução obtida y = 5*exp(sen(t))-4
import numpy as np;
import matplotlib.pyplot as plt;

t = np.linspace(0, 5)
y = 5*np.exp(np.sin(t))-4
plt.plot(t, y)
plt.xlabel('t');
plt.ylabel('y(t)');
```

Obtemos o mesmo gráfico, logo podemos assumir que a solução está correta

Parte 2: *dy/dt = y exp(t)*:

\begin{equation}
y'= ye^{t}
\end{equation}

Transformando para o formato das equações separáveis:

\begin{equation}
g(t) = e^{t}
\end{equation}

\begin{equation}
f(y) = \frac{1}{y}
\end{equation}

Encontrando a solução y sabendo que y0 = 1 e t0 = 0:

\begin{equation}
\int_{0}^{t} e^{t} dt = \int_{1}^{y} \frac{1}{y} dy
\end{equation}

Resolvendo as integrais:

\begin{equation}
e^{t} - 1 = ln(y)
\end{equation}

E, por fim, exponenciando de ambos os lados:

\begin{equation}
y = e^{e^{t}-1}
\end{equation}

Verificando o gráfico para a solução calculada numericamente em python da equação diferencial:

\begin{equation}
y' = ye^{t}
\end{equation}

```python
#Mostrar o gráfico de y'= y exp(t)
import numpy as np;
from scipy.integrate import odeint
import matplotlib.pyplot as plt;

def function(y, t):
  dydt = y*np.exp(t)
  return dydt

y0 = 1
t = np.linspace(0, 5)
Y = odeint(function, y0, t)
plt.plot(t, Y)
plt.xlabel('t');
plt.ylabel('y(t)');
```

Vamos verificar o grafico da solução obtida analiticamente:

\begin{equation}
y = e^{e^{t}-1}
\end{equation}

```python
#Mostrar o gráfico da solução obtida y = exp(exp(t)-1)
import numpy as np;
import matplotlib.pyplot as plt;

t = np.linspace(0, 5)
y = np.exp(np.exp(t)-1)
plt.plot(t, y)
plt.xlabel('t');
plt.ylabel('y(t)');
```

Novamente obtivemos o mesmo gráfico, logo podemos assumir que essa solução também está correta
